```python
data = {
    "salary":[25000,30000,45000,50000,55000,65000,67000,70000,72000]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
#lineplot:
```

```python
plt.plot(df["salary"] , color="red", marker = "o",linestyle = ":",linewidth="2")
plt.grid()
plt.show()
```

```python
# Histogram
plt.hist(df["salary"],bins=10)
plt.show()
```

```python
#boxplot

plt.boxplot(df["salary"])
plt.show()
```

```python
df["dept"] = ['HR','IT','FINANCE','IT','FINANCE','HR','FINANCE','COMPUTER','HR']
df.head()
```

```python
df.info()
```

```python
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels = count.index , autopct = "%1.1f", explode = [0,0.1,0,0])
plt.show()
```

```python
# countplot
plt.bar(count.index,count , color = ["red",'yellow',"green","blue"])
plt.show()

```

```python
#bivariate = numerical-numerical
```

```python
df.head()
```

```python
df["age"]=[25,34,29,27,30,23,25,21,28]
df.head()
```

```python
df.info()
```

```python
#scatter plot:
plt.scatter(df["age"],df["salary"])
plt.show()
```

```python
#lineplot

plt.plot(df["age"],df["salary"],color="red",marker="*",linestyle=":",linewidth="2")
plt.grid()
plt.show()
```

```python
#bar chart:
plt.bar(df["age"],df["salary"],color="red")
plt.show()
```

```python
#bivariate =  numericals - categorical

hr_sal = df[df["dept"] =="HR"]["salary"]
hr_sal
```

```python

it_sal = df[df["dept"] =="IT"]["salary"]
it_sal
```

```python
fin_sal = df[df["dept"] =="FINANCE"]["salary"]
fin_sal
```

```python
plt.boxplot([hr_sal,it_sal,fin_sal],labels=["HR","IT","FINANCE"] )
plt.show()
```

```python
#pie chart:
salary_by_dept = df.groupby("dept")["salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct="%1.2f")
plt.show()
```

```python
#barplot
hr_mean  = sum(hr_sal)/len(hr_sal)
it_mean  = sum(it_sal)/len(it_sal)

plt.bar(["HR","IT"] , [hr_mean , it_mean] , color = ["green","black"])
plt.show()
```

```python
#multivariate analysis:3 numericals column
df["experience"]=[1,2,3,6,8,4,9,5,3]
df.head()
```

```python
#bubble plot:
plt.scatter(df["age"],df["salary"], s=df["experience"]*20,color="skyblue",edgecolors="black")
plt.title("age vs salary vs experience")
plt.xlabel("age")
plt.ylabel("salary")
plt.grid()
plt.show()
```

```python
plt.scatter(df["age"],df["salary"],c= df["dept"].map({"HR":"yellow" ,"IT":"red","FINANCE":"orange", "COMPUTER":"blue"}),label = df["dept"])
plt.xlabel("age")
plt.ylabel("salary")
plt.title("age vs salary vs dept")
plt.grid()
plt.show()
```

```python
color = {"HR":"yellow" ,"IT":"red","FINANCE":"orange", "COMPUTER":"blue"}
for dept,color in color.items():
  df_dept  = df[df["dept"]==dept]
  plt.scatter(df_dept["age"],df_dept["salary"],c = color ,label = dept)
  plt.legend()
```

```python
#object oriented API
```

```python

fig , axs = plt.subplots(1,3,figsize=(10,5))
#lineplot
axs[0].plot(df["age"],df["salary"],color = "red",marker = "*",linewidth="2",markersize =  2)
axs[0].grid()
axs[0].set_title("Lineplot")
axs[0].set_xlabel("age")
axs[0].set_ylabel("salary")
plt.show()


```

```python
#histogram
fig , axs = plt.subplots(1,3,figsize=(10,5))
axs[0].hist(df["salary"],bins = 5,color = "red")
axs[0].set_title("Histogram")
axs[0].set_xlabel("salary")
axs[0].set_ylabel("frequency")
plt.show()
```

```python
#boxplot
fig , axs = plt.subplots(1,3,figsize=(10,5))
axs[0].boxplot(df["salary"])
axs[0].set_title("boxplot")
axs[0].set_xlabel("salary")




plt.show()
```

```python
ax = plt.axes(projection = "3d")
ax.scatter(df["age"],df["salary"],df["experience"])
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_zlabel("Experience")
plt.show()
```
